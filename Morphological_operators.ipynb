```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def load_img():
    blank_img =np.zeros((600,600))
    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(blank_img,text='ABCDE',org=(50,300), fontFace=font,fontScale= 5,color=(255,255,255),thickness=25,lineType=cv2.LINE_AA)
    return blank_img
```

```python
def display_img(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
```

```python
img = load_img();
display_img(img);
```

```python
kernel = np.ones((5,5), dtype=np.uint8)
# Morphological operators are basically kernels used for special effects
# Erosion is used to erode boundries of foreground objects
```

```python
kernel
```

```python
res = cv2.erode(img,kernel, iterations=4)
display_img(res)
```

```python
# Using OPENING is erosion followed by dillation(offset of adding more to the foreground)
# Opening can be useful to reduce BACKGROUND noise (specially in grayscale, for bnw)
img = load_img()
noise = np.random.randint(low=0, high=2,size=(600,600))
noise
```

```python
display_img(noise)
```

```python
# Translating to scale of our image (0 - 255 not 0 - 1) and then combining them
noise = noise * 255
```

```python
noise_img = noise + img
```

```python
display_img(noise_img)
```

```python
opening = cv2.morphologyEx(noise_img,cv2.MORPH_OPEN,kernel)
```

```python
display_img(opening)
```

```python
# Now we are gonna reduce FOREGROUND noise with CLOSING
img = load_img()
black_noise = np.random.randint(low=0,high=2,size=(600,600))
```

```python
black_noise = black_noise * -255
black_noise_img = img + black_noise
black_noise_img[black_noise_img == -255] = 0
display_img(black_noise_img)
```

```python
closing = cv2.morphologyEx(black_noise_img,cv2.MORPH_CLOSE,kernel)
```

```python
display_img(closing)
```

```python
# Morphological Gradient is the difference between the Erosion and dillation of an image
img = load_img()
display_img(img)
```

```python
gradient = cv2.morphologyEx(img,cv2.MORPH_GRADIENT,kernel)
```

```python
display_img(gradient)
```
