```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('PhotosLARC/car_plate.jpg')
```

```python
def display(img):
    fig = plt.figure(figsize=(10,8))
    ax = fig.add_subplot(111)
    correct_color = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    ax.imshow(correct_color)
```

```python
display(img)
```

```python
plate_cascade = cv2.CascadeClassifier('haarcascades/haarcascade_licence_plate_rus_16stages.xml')
```

```python
# Function that detects plates of cars
def detect_plate(img):
    plate_img = img.copy()
    plate_recs = plate_cascade.detectMultiScale(plate_img)
    for(x,y,w,h) in plate_recs:
        cv2.rectangle(plate_img,(x,y),(x + w,y + h), (255,0,0),3)
    return plate_img
```

```python
res = detect_plate(img)
```

```python
display(res)
```

```python
# Function that detects and blurs plate
def detect_and_blur_plate(img):
    plate_img = img.copy()
    roi = img.copy()
    
    plate_recs = plate_cascade.detectMultiScale(plate_img)
    
#   Select ROI and then 
    for (x,y,w,h) in plate_recs:
        roi = roi[y:y+h,x:x+w]
#         Blur it and then blend it into the image
        blurred_roi = cv2.medianBlur(roi,7)
        plate_img[y:y+h,x:x+w] = blurred_roi
    
    return plate_img
```

```python
blurred_plate = detect_and_blur_plate(img)
```

```python
display(blurred_plate)
```
