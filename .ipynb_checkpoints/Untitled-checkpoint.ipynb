```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('PhotosLARC/oneHalfHorizontal.png')
plt.imshow(img)
```

```python
edges = cv2.Canny(image=img, threshold1=127,threshold2=127)
```

```python
plt.imshow(edges)
```

```python
# To improve performance, blur image first
blurred = cv2.blur(img,ksize=(5,5))
med_val = np.median(blurred)
med_val
```

```python
# Set lower threshold to 0 or 70% of the median value, depending on which is greater
lower = int(max(0,0.7*med_val))
# Set upper threshold to 255 or 130% of the median value, depending on which is smaller
upper = int(min(255,1.3*med_val))
```

```python
edges = cv2.Canny(image=blurred, threshold1=lower,threshold2=upper)
plt.imshow(edges)
```
