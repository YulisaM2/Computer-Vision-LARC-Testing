```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def load_img():
    img = cv2.imread('../PhotosLARC/horizontal.png').astype(np.float32) / 255
    img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    return img
```

```python
def display_img(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img)
```

```python
i = load_img()
```

```python
display_img(i)
```

<!-- Using gamma to control level of brightness (gamma lower than 1 is brighter, larger than 1 are darker) -->

```python
gamma = 1/4
```

```python
result = np.power(i,gamma)
display_img(result)
```

```python
img = load_img()
font = cv2.FONT_HERSHEY_COMPLEX
cv2.putText(img,text="stacks",org=(700,1000),fontFace=font,fontScale=30,color=(255,0,0),thickness=4)
display_img(img)
```

```python
# Moving into blurring and smoothing by creating kernel of 5 x 5
kernel = np.ones(shape=(5,5),dtype=np.float32)/25
```

```python
kernel
```

```python
dst = cv2.filter2D(img,-1,kernel)
display_img(dst)
```

```python
# Reseting values
img = load_img()
font = cv2.FONT_HERSHEY_COMPLEX
cv2.putText(img,text="stacks",org=(700,1000),fontFace=font,fontScale=30,color=(255,0,0),thickness=4)
```

```python
blurred = cv2.blur(img,ksize=(5,5))
display_img(blurred)
```

```python
# Reseting values
img = load_img()
font = cv2.FONT_HERSHEY_COMPLEX
cv2.putText(img,text="stacks",org=(700,1000),fontFace=font,fontScale=30,color=(255,0,0),thickness=4)
```

```python
blurredGauss = cv2.GaussianBlur(img,(5,5),10)
display_img(blurredGauss)
```

```python
# Reseting values
img = load_img()
font = cv2.FONT_HERSHEY_COMPLEX
cv2.putText(img,text="stacks",org=(700,1000),fontFace=font,fontScale=30,color=(255,0,0),thickness=4)
```

```python
# This method actually helps reduce noise too
blurredMedian = cv2.medianBlur(img,5)
display_img(blurredMedian)
```

```python
#  Reseting values
img = load_img()
font = cv2.FONT_HERSHEY_COMPLEX
cv2.putText(img,text="stacks",org=(700,1000),fontFace=font,fontScale=30,color=(255,0,0),thickness=4)

# Bilateral filters can also help blur AND reduce noise (WHILE KEEPING EDGES SHARP!)
blurredBi = cv2.bilateralFilter(img,9,75,75)
display_img(blurredBi)
```
