```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def display_img(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
```

```python
img = cv2.imread('../PhotosLARC/horizontal.png',0)
```

```python
display_img(img)
```

```python
# Gradient for plane x 
sobelX = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5)
```

```python
display_img(sobelX)
```

```python
# Gradient for plane y
sobelY = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=5)
display_img(sobelY)
```

```python
# Laplace operator is used to combine the result of the second partial derivatives in x and y
laplace = cv2.Laplacian(img,cv2.CV_64F)
display_img(laplace)
```

```python
# You can also try and blend sobelX and sobelY to get a better result
```

```python
blended = cv2.addWeighted(src1=sobelX,alpha=0.5,src2=sobelY,beta=0.5,gamma=0)
```

```python
display_img(blended)
```

```python
ret,th1 = cv2.threshold(blended,100,255,cv2.THRESH_BINARY)
display_img(th1)
```

```python
kernel = np.ones((4,4),np.uint8)
gradient = cv2.morphologyEx(blended,cv2.MORPH_GRADIENT,kernel)
display_img(gradient)
```
