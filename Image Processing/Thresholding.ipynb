```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Read image in grayscale image (thanks to param=0)
img = cv2.imread('../PhotosLARC/front.jpg',0)
# img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img,cmap="gray")
```

```python
ret,thresh1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
plt.imshow(thresh1,cmap="gray")
```

```python
ret,threshInv = cv2.threshold(img,127,255,cv2.THRESH_BINARY_INV)
plt.imshow(threshInv,cmap="gray")
```

```python
# If source val is > threshold, it becomes treshold value, otherwise if it is less than threshold, source is kept
ret,threshTrunc = cv2.threshold(img,127,255,cv2.THRESH_TRUNC)
plt.imshow(threshTrunc,cmap="gray")
```

```python
# Use neighbor pixels to adapt the threshold of a region 
threshAdapt = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,8)
```

```python
plt.imshow(threshAdapt,cmap="gray")
```
